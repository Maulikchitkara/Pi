```python
#monte-carlo
import random as ran 

N = 1000000
inside_circle = 0

for _ in range(N):
    x= ran.random()
    y= ran.random()

    if x*x + y*y <=1:
        inside_circle += 1

pi = 4*inside_circle/N

print(pi)
```

```python
N = 5000

inside_x = []
inside_y = []
outside_x = []
outside_y = []

inside = 0

plt.ion()
fig, ax = plt.subplots()

for i in range(1, N+1):

    x = ran.random()
    y = ran.random()

    if x*x + y*y <= 1:
        inside += 1
        inside_x.append(x)
        inside_y.append(y)
    else:
        outside_x.append(x)
        outside_y.append(y)

    if i % 2000 == 0: 

        ax.clear()

        ax.scatter(inside_x, inside_y, s=2)
        ax.scatter(outside_x, outside_y, s=2)

        pi_estimate = 4 * inside / i

        ax.set_title(f"π ≈ {pi_estimate}  after {i} points")

        ax.set_xlim(0,1)
        ax.set_ylim(0,1)

        plt.pause(0.01)

plt.ioff()
plt.show()
```

```python
#Leibniz Series

pi = 0 
N=10000

for k in range(N):
    pi+= 4*(-1)**k / (2*k + 1)
    k+=1

print(pi)
```

```python
#Machin
pi_1= 16*Decimal(np.arctan(0.2)) - 5*Decimal(np.arctan(1/239))
print(pi_1)

def arctan(x):
    arctan=0
    for k in range(N):
        arctan+= ((-1)**k)*(x**(2*k+1))/(2*k+1)
        k+=1
    return arctan

pi=16*Decimal(arctan(0.2)) - 5*Decimal(arctan(1/239))

print(pi)
```
